## Anàlisi Trajectòria

```python
u=mda.Universe('adk.psf','unfolding.dcd')
```

```python
print(len(u.trajectory)) #concorda amb el numero de frames del VMD
```

```python
print(u.trajectory)
```

```python
print(u.trajectory.time)
```

```python
for ts in u.trajectory: #de la trajectrotia agafar el temps i radir de gir en angstroms
    time = u.trajectory.time
    rgyr = u.atoms.radius_of_gyration()
    print(f"Frame: {ts.frame:3d}, Time: {time:4.0f} ps, Rgyr: {rgyr:.4f} A")
```

```python
rgyr = []
time = []
protein = u.select_atoms("protein")
for ts in u.trajectory:
    time.append(u.trajectory.time)
    rgyr.append(protein.radius_of_gyration())
```

```python
plt.plot(time,rgyr)
plt.xlabel('Time (ps)')
plt.ylabel('Rg (A)')
plt.show()
```

**save into file**

```python
fitxer=open("Rg.dat","w")
```

```python
for i in range(len(time)):
    fitxer.write(f'{round(time[i],1)} {round(rgyr[i],2)}\n')
```
